#### Criação de gráficos e visualizações de dados em geral.

Vamos utilizar o famoso dataset de preços de casas da Califórnia - EUA.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("preco_casas.csv")
```

```python
df.head()
```

Utilizando o Seaborn para visualizar a Dispersão entre as variáveis salário_mediano e preco_mediano_das_casas

__Seaborn é uma biblioteca python para gráficos estatísticos.__

```python
sea.jointplot(df.salario_mediano, df.preco_mediano_das_casas,df, kind='hist')
```

O joinplot cria vários gráficos em um só. Dois histogramas em cada eixo, dispersão dos dados já incluindo a 
densidade dos dados. Podemos ver uma maior densidade entre os salários até 6.000 e valor das casas 300.000.

```python
sea.jointplot(df.salario_mediano, df.preco_mediano_das_casas,df, kind='hex', color="#4CB391")
```

Focando a densidade entre as duas variáveis.

```python
sea.jointplot(df.salario_mediano, df.preco_mediano_das_casas,df, kind='kde')
```

```python
sea.jointplot(df.salario_mediano, df.preco_mediano_das_casas,df, kind='reg', color="#4CB391")
```

Acima: Gráfico aplicando uma linha de regressão e distribuição dos dados.

##### Abaixo vamos utilizar o matplotlib. Biblioteca para visualização de dados.

```python

plt.figure(figsize=[12,10])
plt.scatter(df.longitude, df.latitude, c = df.preco_mediano_das_casas, s = df.populacao / 50, cmap = plt.cm.rainbow)
plt.title("Distribuição do preço x População", fontsize = 12)
plt.xlabel("Longitude", fontsize = 14, fontweight = 'bold')
plt.ylabel("Latitude", fontsize = 14, fontweight = 'bold')
plt.colorbar()
plt.grid()
plt.show()
```

Observe os círculos maiores: Indicam áreas mais populosas. Áreas vermelhas indicam regiões com preços mais altos
e estão mais próximas ao mar.
